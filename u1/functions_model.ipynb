# Functions and Model

functions are everywhere, like quite literally everywhere, whats cool about them
is they take a number and ouput another. Yes in a way they kind of relate numbers
together.

Now here is the right time to introduce the notion of *domain* and *range*. Now
Domain is all the valid values you can give as a input to the function where-as
range is all the values a function can take. for example $y= f(x)$ where $f(x)= x + 2$
now the above funcion is very intutive in the sense that you can give in a value of
$x$ which by the way is called as the independent variable and $y$ is called
as the dependent variable.

> It is called as the independent variable simply because its value is indipendent
of anything where-as the value of $y$ is dependent on the value of $x$.

Before we calculate the value of the above function let us quickly plot the above

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(-10, 10, 400)
y = x + 2 #our function to plot

plt.plot(x, y, label="f(x) = x + 2")
plt.axhline(0, color="black", linewidth=0.5)  # x-axis
plt.axvline(0, color="black", linewidth=0.5)  # y-axis
plt.legend()
plt.grid(True)
plt.show()
```

So interstingly enough we see that both the domain set and the range set can be 
anything between $(-\infty, +\infty)$. Now let us step up and understand another
fucntion  before we go and see some different types of functions. consider $y= |x|$.
We will write our own *abs* function to calculate the absolute to warm up. below
is the formal definition of *abs*.

$$
|x| =
\begin{cases}
   x, & \text{if } x \geq 0 \\
  -x, & \text{if } x < 0
\end{cases}
$$

The above is also called as the piece-wise definition of the function. And from 
the plot below you can make out that the domain is $(-\infty, \infty)$ but the 
range is $[0, \infty)$.


```python

# you can also use the np.absolute value as well
def mod(x):
    if x < 0:
        return -1 * x
    else:
        return x

x= np.linspace(-100, 100, 4000)
y= list(map(mod, x))

plt.plot(x, y, label="f(x) = x + 2")
plt.axhline(0, color="black", linewidth=0.5)  # x-axis
plt.axvline(0, color="black", linewidth=0.5)  # y-axis
plt.legend()
plt.grid(True)
plt.show()
```
